# Bitcoin Market Analysis

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np


df = pd.read_csv('BTC_all.csv')
df.head()
```

```python
YEAR_LENGTH = 365
year_names = ['2014', '2015', '2016', '2017', '2018', '2019']
years = {}
start_idx = 93
for year_name in year_names:
    start = df[df.Date == f'{year_name}-01-01'].index.item()
    end = df[df.Date == f'{year_name}-12-31'].index.item()
    prices = df.loc[start:end].get('Closing Price (USD)').values
    years[year_name] = prices
    

plt.figure(figsize=(15, 5))
plt.plot(years['2019'])
plt.show()
```
